```python
# importing the libraries

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Data collection and data processing
# loading dataset to the pandas dataframe

sonar_data=pd.read_csv('/content/Copy of sonar data.csv',header=None)
```

```python
#number of rows and coloumn
sonar_data.shape
```

```python
# statastical measure of the data
sonar_data.describe()
```

```python
#counting how many rock and how many mine data

sonar_data[60].value_counts()
```

```python
# seperating data and labels

X=sonar_data.drop(columns=60,axis=1) # here axis 1 means coloumn and axis 0 means row
Y=sonar_data[60]
```

```python
# spliting this data into training and test data

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
#loading the logistic regression model

model=LogisticRegression()
```

```python
# training logistic regression model using our data

model.fit(X_train,Y_train)
```

```python
# model evaluation

# accuracy on training data

X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)

print('Accuracy on training data:',training_data_accuracy)
```

```python
# accuracy on test data

X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)

print('Accuracy on test data:',test_data_accuracy)
```
